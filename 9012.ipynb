```python
n = int(input())

for i in range(n):
    st = input()
    
    a = st.count('(')
    b = st.count(')')
    
    if a == b:
        print('YES')
    else:
        print('NO')
```

```python
n = int(input())

for i in range(n):
    st = input()
    
    c = 0
    
    for i in st:
        if i == '(':
            c += 1
        else:
            c -= 1
            
        if c < 0:
            break
            
    if c == 0:
        print('YES')
    else:
        print('NO')
```
