```python
board = [["blue", "red", "orange", "red"], ["red", "red", "blue", "orange"], ["blue", "orange", "red", "red"], ["orange", "orange", "red", "blue"]]
h = 1
w = 1
```

```python
row = len(board)
col = len(board[0])

target_color = board[h][w]

direction = [(1,0),(-1,0),(0,1),(0,-1)]

count = 0

for dx, dy in direction:
    nx, ny = h + dx, w + dy
    if 0 <=  nx < row and 0 <= ny < col and board[nx][ny] == target_color:
        count += 1
        
        
count
```

```python
def solution(board, h, w):
    # 주어진 칸의 색깔
    target_color = board[h][w]
    # 같은 색깔의 인접 칸 개수를 셀 변수
    count = 0

    # 인접한 네 방향의 좌표 (위, 아래, 왼쪽, 오른쪽)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    for dh, dw in directions:
        nh, nw = h + dh, w + dw
        # 인접한 칸이 보드판의 범위 내에 있고, 같은 색인지 확인
        if 0 <= nh < len(board) and 0 <= nw < len(board[0]) and board[nh][nw] == target_color:
            count += 1

    return count
```
