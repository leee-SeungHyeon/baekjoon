```python
while True:
    m = int(input())
    if m < 0:
        break
    nums = []
    
    for i in range(1,m+1):
        if m%i==0:
            nums.append(i)

    if m == sum(nums[:-1]):
        print(str(m)+' = '+' + '.join(map(str, nums[:-1])))
    else:
        print('{} is NOT perfect.'.format(m))
```
