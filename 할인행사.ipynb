```python
def solution(want, number, discount):
    
    want_book = {}

    for w, n in zip(want, number):
        want_book[w] = n
        
    count = 0

    for start in range(len(discount) - 10 + 1):
        want_book_ = want_book.copy()
        for j in range(start,10+start):
            if discount[j] in want_book_ and want_book_[discount[j]] != 0:
                want_book_[discount[j]] -= 1

        if all(value == 0 for value in want_book_.values()):
            count += 1
    
    return count
```
