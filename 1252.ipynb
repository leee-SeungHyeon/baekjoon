```python
num1, num2 = map(str, input().split())

a = 0
for i, inx in enumerate(reversed(num1)):
    if int(inx) != 0:
            a += 2**i
            
b = 0
for i, inx in enumerate(reversed(num2)):
    if int(inx) != 0:
            b += 2**i

c = a+b

result = []

while True:
    num = c % 2
    c = c//2
    result.append(num)
    
    if c < 2:
        result.append(c)
        break

result.reverse()

for i in result:
    print(i,end='')
```
