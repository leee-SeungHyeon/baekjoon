```python
s = input()

aList = [chr(i) for i in range(ord('a'),ord('z')+1)]

for i in s:
    for j in range(len(aList)):
        if i == aList[j]:
            aList[j] = s.index(i)

for z in range(len(aList)):
    if str(aList[z]).isalpha():
        aList[z] = -1

for i in aList:
    print(i, end=' ')
```

- 내코드는 맞기는 했지만, for문도 많고 너무 복잡하게 짰다.
- 찾아보니 find 함수를 쓰면 복잡함이 한방에 해결되는것 같다.
- 아직 연습이 부족하다.

```python
s = input()

aList = [chr(i) for i in range(ord('a'),ord('z')+1)]

for x in aList:
    print(s.find(x), end=' ')
```
