```python
s = input()

aList = [chr(i) for i in range(ord('a'),ord('z')+1)]

for i in s:
    for j in range(len(aList)):
        if i == aList[j]:
            aList[j] = s.index(i)

for z in range(len(aList)):
    if str(aList[z]).isalpha():
        aList[z] = -1

for i in aList:
    print(i, end=' ')
```
