```python
def solution(l, r):
    answer = []
    
    for i in range(l,r+1):
        if all(num in ['0','5'] for num in str(i)):
            answer.append(i)
            
    if len(answer) == 0:
        answer.append(-1)

    return answer
```

 - all() 함수는 주어진 조건식을 리스트 내의 모든 요소에 적용하고, 모든 조건이 True인 경우에만 True를 반환합니다. 그렇지 않으면 False를 반환

 - 따라서 all(num in ['0','5'] for num in str(i))는 주어진 숫자의 각 자릿수가 "0" 또는 "5"로만 이루어져 있는지를 검사하며, 모든 자릿수가 이 조건을 만족해야 True가 되고, 그렇지 않으면 False
