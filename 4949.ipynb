```python
while True:
    words = input()
    
    if words == '.':
        break
    
    answer = []
    
    for word in words:
        if word == '(' or word == '[':
            answer.append(word)
        
        elif word == ')':
            if len(answer) != 0 and answer[-1] == '(':
                answer.pop()
            else:
                answer.append(word)
                break
        
        elif word == ']':
            if len(answer) != 0 and answer[-1] == '[':
                answer.pop()
            else:
                answer.append(word)
                break
                    
    if len(answer) == 0:
        print('yes')
    else:
        print('no')
        
```
