```python
def is_prime(num):
    if num < 2:
        return False
    else:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True

n = int(input())

for _ in range(n):
    num = int(input()) 
    while True:
        if is_prime(num):
            print(num)
            break
        else:
            num+=1
```
