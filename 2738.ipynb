```python
a, b = [], []

N, M = map(int, input().split())

for i in range(N):
    r = list(map(int, input().split()))
    a.append(r)
    
for j in range(N):
    r = list(map(int, input().split()))
    b.append(r)
    
for i in range(N):
    for j in range(M):
        print(a[i][j] + b[i][j], end=' ')
    print()
```
