```python
n = int(input())

words_list = []

for _ in range(n):
    words_list.append(input())
    
words_list = set(words_list)
words_list = list(words_list)

words_list.sort(key=lambda x: (len(x),x))

for i in words_list:
    print(i)
```

- tuple을 비교 원리를 제대로 안 것은 처음이다.
- words_list.sort(key=lambda x: (len(x), x)) 작동원리
    1. len(x)를 이용하여 단어의 길이를 계산
    2. (len(x), x)를 반환하여 각 단어를 길이와 단어 자체를 tuple로 묶음
    3. sort() 메서드는 이 tuple을 이용하여 정렬
    4. tuple의 첫 번째 원소인 길이를 기준으로 먼저 정렬을 수행하고,
    5. 단어들의 길이가 같은 경우에는 tuple의 두 번째 원소인 단어 자체를 비교하여 사전순으로 정렬
