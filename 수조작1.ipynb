```python
def solution(n, control):

    w = control.count('w')
    a = control.count('a')
    s = control.count('s')
    d = control.count('d')
    
    n = n + (w * 1) + (a * -10) + (s * -1) + (d * 10)
    return n
```

```python
def solution(n, control):

    for i in control:
        if i == 'w':
            n += 1
        elif i == 'a':
            n -= 10
        elif i == 's':
            n -= 1
        else:
            n += 10
    return n
```
