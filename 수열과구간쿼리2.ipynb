```python
def solution(arr, queries):
    answer = []
    
    for i in range(len(queries)):
        min_num = 1000000
        for j in range(queries[i][0],queries[i][1]+1):
            if queries[i][2] < arr[j]:
                min_num = min(min_num,arr[j])
        if min_num == 1000000:
            min_num = -1
        answer.append(min_num)
    return answer
```
