```python
def round_num(n):
    return int(n) + (1 if n - int(n) >= 0.5 else 0) 

n = int(input())
if not n:
    print(0)
else:
    score = []

    for _ in range(n):
        mm = int(input())
        score.append(mm)
    
    score.sort() 
    tr = round_num(n*0.15)

    new_score = score[tr:-tr]

    print(round(sum(new_score)/len(new_score)))    
    
```

```python
print(round(1.5))
print(round(2.5))
print(round(3.5))
print(round(4.5))
```

round() 함수는 소수를 가까운 정수로 반올림하는 함수입니다. 그러나, 소수가 정확히 중간에 위치하는 경우, 즉 소수 부분이 0.5일 때, round() 함수는 가장 가까운 짝수 정수로 반올림을 수행합니다.

따라서, round(1.5)는 2로 반올림되고, round(2.5)도 2로 반올림됩니다. 이는 반올림의 규칙으로서 소수 부분이 정확히 0.5일 때 가장 가까운 짝수 정수로 반올림되는 것입니다.

그래서 반올림하는 함수를 다시 구현해주었습니다.
