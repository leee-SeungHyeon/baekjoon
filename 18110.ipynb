```python
def round_num(n):
    return int(n) + (1 if n - int(n) >= 0.5 else 0) 

n = int(input())
if not n:
    print(0)
else:
    score = []

    for _ in range(n):
        mm = int(input())
        score.append(mm)
    
    score.sort() 
    tr = round_num(n*0.15)

    new_score = score[tr:-tr]

    print(round(sum(new_score)/len(new_score)))    
    
```
