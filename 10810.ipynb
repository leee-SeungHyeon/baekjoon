```python
a, b = map(int, input().split()) #   a    b

busket = [0]*a #  0   

for _ in range(b): #   b 
    i, j, k = map(int, input().split())
    for n in range(i,j+1):
        busket[n-1] = k
        
for i in range(a):
    print(busket[i], end=' ')
```
