```python
a, b = map(int, input().split()) #바구니와 공 번호인 a와 몇번 넣을지 횟수인 b

busket = [0]*a #바스켓에는 무조건 0이 들어가 있어야 함

for _ in range(b): #공 넣는 횟수 b만큼 반복
    i, j, k = map(int, input().split())
    for n in range(i,j+1):
        busket[n-1] = k
        
for i in range(a):
    print(busket[i], end=' ')
```
