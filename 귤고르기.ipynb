```python
def solution(k, tangerine):
    frute = {}
    
    # 과일 갯수 사전 만들기
    for i in tangerine:
        if i in frute:
            frute[i] += 1
        else:
            frute[i] = 1
    
    # values값을 기준으로 재정렬
    sorted_frute = dict(sorted(frute.items(), key=lambda item: item[1], reverse=True))
    
    answer = 0
    
    # 과일 갯수 사전의 수만큼 k에서 빼주기
    for i in sorted_frute:
        if k <= 0:
            return answer
        else:
            k -= sorted_frute[i]
            answer += 1
    
    return answer
```
