```python
def solution(k, tangerine):
    frute = {}
    
    #    
    for i in tangerine:
        if i in frute:
            frute[i] += 1
        else:
            frute[i] = 1
    
    # values  
    sorted_frute = dict(sorted(frute.items(), key=lambda item: item[1], reverse=True))
    
    answer = 0
    
    #     k 
    for i in sorted_frute:
        if k <= 0:
            return answer
        else:
            k -= sorted_frute[i]
            answer += 1
    
    return answer
```
