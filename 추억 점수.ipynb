```python
def solution(name, yearning, photo):

    dd = {}

    for na, ye in zip(name, yearning):
        dd[na] = ye

    sum_list = []
    sum = 0
    for i in photo:
        for z in i:
            if z in dd:
                sum += dd[z]

        sum_list.append(sum)
        sum = 0   
    return sum_list
```
