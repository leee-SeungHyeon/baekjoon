```python
while True:
    
    a,b,c = map(int, input().split())
    
    if a == b == c == 0:
        break
    
    num = [a,b,c]
    
    max1 = max(num)
    num.remove(max(num))
    
    if max1 >= sum(num):
        print('Invalid')
        continue
    
    if a == b == c:
        print('Equilateral')
    elif (a == b) or (a == c) or (b == c):
        print('Isosceles')
    elif a != b != c:
        print('Scalene')
```
