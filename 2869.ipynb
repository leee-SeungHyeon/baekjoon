```python
a,b,v = map(int, input().split())

c = 0

while True:
    v -= a
    if v == 0 or v < 0:
        c+=1
        break
    else:
        v+=b
        c+=1

print(c)
    
```

```python
import math

a,b,v = map(int, input().split())

print(math.ceil(((v-b)/(a-b))))

```
