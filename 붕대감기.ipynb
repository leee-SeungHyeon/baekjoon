```python
def solution(bandage, health, attacks):
    
    t, x, y = bandage
    max_hp = health
    end_time = attacks[-1][0]
    attacks = {attack[0]: attack[1] for attack in attacks}
    
    band = 0
    current_health = max_hp
    
    for i in range(end_time+1):
        if i in attacks:
            band = 0 
            current_health -= attacks[i]
            
            if current_health <= 0:
                return -1
            
        else: 
            band += 1
            current_health += x

            if band == t:
                current_health += y
                band = 0
                
            if current_health >= max_hp:
                current_health = max_hp
    
    return current_health
```
