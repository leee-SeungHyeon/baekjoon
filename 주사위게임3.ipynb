```python
def solution(a, b, c, d):

    bbb_dic = dict()
    bbb_list = list([a,b,c,d])
    for i in bbb_list:
        bbb_dic[i] = bbb_list.count(i)

    bbb_dic =dict(sorted(bbb_dic.items(), key = lambda x : x[1], reverse=True)) #이것을 해주는것이 핵심
    
    bbb_dic_keys = list(bbb_dic.keys())

    if len(bbb_dic) == 1:
        return bbb_dic_keys[0] * 1111

    elif len(bbb_dic) == 2 and bbb_dic[bbb_dic_keys[0]] == 2:
        return ((bbb_dic_keys[0] + bbb_dic_keys[1]) * abs(bbb_dic_keys[0] - bbb_dic_keys[1]))

    elif len(bbb_dic) == 2 and bbb_dic[bbb_dic_keys[0]] != 2:
        return ((10 * bbb_dic_keys[0] + bbb_dic_keys[1])**2)

    elif len(bbb_dic) == 3:
        return bbb_dic_keys[1] * bbb_dic_keys[2]

    else:
        return min(bbb_dic_keys)
```
