```python
n, m = map(int, input().split())

busket = [i for i in range(n+1)]

for _ in range(m):
    i, j = map(int, input().split())
    
    busket[i:j+1] = reversed(busket[i:j+1])
busket.remove(busket[0])

for i in busket:
    print(i, end=' ')
        
```

```python
#  
#    
#basket[i-1:j][::-1]     

n,m = map(int, input().split())
basket = [i+1 for i in range(n)]
for _ in range(m):
    i, j = map(int, input().split())
    basket = basket[:i-1] + basket[i-1:j][::-1] + basket[j:]
    print(basket)
for i in basket:
    print(i, end = ' ')
```
