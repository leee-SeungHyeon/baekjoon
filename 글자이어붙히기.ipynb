```python
def solution(my_string, index_list):
    dicc = {}

    for i, v in enumerate(my_string):

        dicc[i] = v
        
    answer = ''
    for i in index_list:
        answer += dicc[i]
        
    return answer
```

```python
my_string = "cvsgiorszzzmrpaqpe"

dicc = {}

for i, v in enumerate(my_string):

    dicc[i] = v

index_list = [16, 6, 5, 3, 12, 14, 11, 11, 17, 12, 7]
result = ''
for i in index_list:
    result += dicc[i]
```
