```python
def solution(progresses, speeds):
    
    answer = []
    d_day = []
    
    for i, j in zip(progresses, speeds):
        if (100 - i)%j == 0:
            zz = (100 - i)//j
        else:
            zz = (100 - i)//j+1
    
        d_day.append(zz)
    
    n = len(progresses)
    
    now = 0
    for i in range(1, len(progresses)):
        if d_day[i] > d_day[now]:
            answer.append(i-now)
            now = i
    answer.append(n-now)
    
    
    return answer
```
