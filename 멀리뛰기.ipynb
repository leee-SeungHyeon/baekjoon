```python
def solution(n):
    '''
    1일때 1가지
    2일떄 2가지
    3일때 111 21 12 3가지
    4일때 1111 211 121 112 22 5가지
    5일때 11111 2111 1211 1121 1112 221 122 212 8가지
    '''
    if n==1:
        return 1
    elif n == 2:
        return 2
    
    else:
        dp = [0] * (n + 1)
        dp[1] = 1
        dp[2] = 2

        for i in range(3,n+1):
            dp[i] = (dp[i-2] + dp[i-1]) % 1234567

        return dp[-1]
```
