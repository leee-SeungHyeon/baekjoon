```python
from collections import deque
# 해당 문제는 그리디로 풀어야함

def solution(people, limit):
    
    people = deque(sorted(people, reverse= True))
    
    answer = 0
    
    while len(people) > 1:
        if people[0] + people[-1] <= limit: # limit를 안넘을 시 가장 무거운 사람과 가장 가벼운 사람 동시에 제거
            people.pop()
            people.popleft()
            answer += 1
        else:
            people.popleft() # limit를 넘을 시 가장 무거운 사람만 제거 
            answer += 1
            
    if len(people) == 1:
        answer += 1
        
    return answer
```
