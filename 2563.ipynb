```python
whitebord = [[0]*100 for i in range(100)]

n = int(input())
sum11 = 0
for _ in range(n):
    x, y = map(int, input().split())
    for i in range(10): #가로
        for j in range(10): #세로
            whitebord[x-1+i][y-1+j] = 1
            
for i in range(len(whitebord)):
    for j in range(len(whitebord[i])):
        sum11 += whitebord[i][j]
        
print(sum11)
```
