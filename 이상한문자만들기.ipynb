```python
def solution(s):
    result = []
    for i in s.split(' '):
        new_word = ''
        for j in range(len(i)):
            if j % 2 == 0:
                new_word += i[j].upper()

            else:
                new_word += i[j].lower()

        result.append(new_word)
    return ' '.join(result)
```
