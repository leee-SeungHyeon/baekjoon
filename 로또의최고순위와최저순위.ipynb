```python
def solution(lottos, win_nums):
    
    winner = {'6':1, '5':2, '4':3, '3':4, '2':5,'1': 6,'0':6}
    
    win = len(set(lottos)&set(win_nums))
    
    zero = lottos.count(0)
        
    # return aa
    return [winner[str(win+zero)], winner[str(win)]]
```
