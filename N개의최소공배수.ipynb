```python
from math import gcd

def solution(arr):

    '''
    gcd(answer,num) : 이렇게 하면 두 수의 최대공약수를 구할 수 있음
    answer * num // gcd(answer,num) : 두 수의 곱 / 최대 공약수를 하면 최소공배수를 구할 수 있음
    '''
    
    answer = arr[0]
    
    for num in arr[1:]:
        answer = answer * num // gcd(answer,num)
    
    return answer
```
