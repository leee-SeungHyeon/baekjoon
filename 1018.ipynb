```python
m , n = map(int, input().split())

maxtrix = []
result = []

for _ in range(m):
    maxtrix.append(input())


for i in range(m-7):
    for j in range(n-7):#8*8  , -7
        white = 0
        black = 0
 
        for a in range(i, i+8):
            for b in range(j, j+8):
                if (a + b) % 2 == 0: # 
                    if maxtrix[a][b] != 'B':# black ,  white
                        white += 1 # white  
                    if maxtrix[a][b] != 'W': 
                        black += 1
                else: #
                    if maxtrix[a][b] != 'W':
                        white += 1
                    if maxtrix[a][b] != 'B':
                        black += 1
 
        result.append(white)
        result.append(black)
 
print(min(result))    
```
