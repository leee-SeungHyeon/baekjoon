```python
m , n = map(int, input().split())

maxtrix = []
result = []

for _ in range(m):
    maxtrix.append(input())


for i in range(m-7):
    for j in range(n-7):#8*8로 자르기 위해, -7해준다
        white = 0
        black = 0
 
        for a in range(i, i+8):
            for b in range(j, j+8):
                if (a + b) % 2 == 0: # 짝수이면
                    if maxtrix[a][b] != 'B':# black이 아니면, 즉 white이면
                        white += 1 # white로 칠하는 갯수
                    if maxtrix[a][b] != 'W': 
                        black += 1
                else: #홀수이면
                    if maxtrix[a][b] != 'W':
                        white += 1
                    if maxtrix[a][b] != 'B':
                        black += 1
 
        result.append(white)
        result.append(black)
 
print(min(result))    
```
