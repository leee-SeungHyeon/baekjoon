```python
def solution(left, right):

    sum = 0
    
    for i in range(left,right+1):
        result = []
        for j in range(1,i+1):
            if i % j == 0:
                result.append(j)
        if len(result) % 2 == 0:
            sum += i
        else:
            sum -= i

    return sum
```
