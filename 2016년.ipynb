```python
def solution(a, b):
    # 각 월의 일 수, 윤년이므로 2월은 29일
    month_days = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    # 요일 배열, 1월 1일은 금요일(FRI)이므로 인덱스 0은 FRI로 시작
    days_of_week = ["FRI", "SAT", "SUN", "MON", "TUE", "WED", "THU"]
    
    # 2016년 1월 1일부터 a월 b일까지 총 일 수 계산
    total_days = sum(month_days[:a-1]) + b - 1  
    
    # 총 일 수를 7로 나눈 나머지를 이용하여 요일 계산
    return days_of_week[total_days % 7]
```
