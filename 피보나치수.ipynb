```python
def solution(n):
    if n==1 or n ==2:
        return 1
    else:
        return (solution(n-1) + solution(n-2)) % 1234567
```

```python
def solution(n):
    pre = 0
    current = 1
    
    for i in range(2,n+1):
        current, pre = current + pre, current
        
    return current % 1234567
```
