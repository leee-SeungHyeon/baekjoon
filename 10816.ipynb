```python
n = int(input())

nn1 = list(map(int, input().split()))

m = int(input())

mm1 = list(map(int, input().split()))

n_list = dict()

for i in nn1:
    n_list[i] = 0
    
for i in nn1:
    if i in n_list:
        n_list[i] += 1
    
for i in mm1:
    if i in n_list:
        print(n_list[i], end= ' ')
    else:
        print(0, end=' ')
```
