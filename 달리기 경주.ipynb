```python
players = ["mumu", "soe", "poe", "kai", "mine"]
callings = ["kai", "kai", "mine", "mine"]
```

```python
player_lank = {name:lank for lank, name in enumerate(players)}
```

```python
def solution(players, callings):
    
    player_lank = {name:lank for lank, name in enumerate(players)}
    
    for call in callings:
        current_lank = player_lank[call]
        
        player_lank[call] -= 1
        player_lank[players[current_lank-1]] += 1
        
        players[current_lank-1], players[current_lank] = call, players[current_lank-1]
    
    return players
```
