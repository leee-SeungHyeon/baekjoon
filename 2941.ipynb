```python
S = input()

count = 0
croatia = ['c=','c-','dz=','d-','lj','nj','s=','z=']
print(S)
for i in croatia:
    if i in S:
        print(i)
        S = S.replace(i,'')
        print(S)
        count+=1
c = len(S)
print(count+c)

# if 'c=' in s:
#     count+=1
# else
```

- 이 문제에서는 replace로 없애면 되겠다라고 했는데, 없애고 나니 위 문제처럼 크로아티아 단어가 생겨서 의미없는 문자로 치환후 마지막에 글자 길이를 세는것으로 방향을 바꿨음
- 생각했던것이 풀리니 짜릿했음
- 다만 적절한 함수 응용을 하려면 더 연습이 필요할것으로 생각됨(인터넷으로 아이디어 힌트보는것을 줄이자!)

```python
S = input()

croatia = ['c=','c-','dz=','d-','lj','nj','s=','z=']

for i in croatia:
    if i in S:
        S = S.replace(i,'q')
print(len(S))
```

```python
#인터넷에 나와있던 풀이

croatia = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']
word = input()

for i in croatia :
    word = word.replace(i, '*')  # input 변수와 동일한 이름의 변수
print(len(word))
```
