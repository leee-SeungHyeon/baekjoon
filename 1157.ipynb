```python
s = input().upper()

unuigue_s = list(set(s))

word_list = []

for i in unuigue_s:
    count = s.count(i)
    word_list.append(count)

if word_list.count(max(word_list)) >=2:
    print('?')
else:
    print(unuigue_s[word_list.index(max(word_list))])
```
