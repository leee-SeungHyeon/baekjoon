```python
n = int(input())

for _ in range(n):
    a, b = map(int, input().split())
    
    for i in range(max(a,b), (a*b)+1):
        if i%a==0 and i%b==0:
            print(i)
            break
```

- 위 코드는 시간초과가 난다
- 아마도 for문때문에 시간복잡도가 많아져서 그런것 같다고 생각된다.

```python
T = int(input())
for _ in range(T):
    A, B = map(int, input().split())
    result = A*B

    while B>0:
        A,B = B, A%B

    print(result//A)
```

- 이것은 찾아본 코드이다.
- 유클리드 호제법을 이용하여 최소공배수를 찾는 과정이다.

Step 1: 두 수의 최대공약수를 구합니다.

- 6과 10의 최대공약수를 구하기 위해 다음과 같이 계산합니다.

- 10을 6으로 나누면 4가 나머지가 됩니다.
- 6을 4로 나누면 2가 나머지가 됩니다.
- 4를 2로 나누면 0이 나머지가 됩니다.
- 따라서 6과 10의 최대공약수는 2입니다.

Step 2: 두 수의 최소공배수를 구합니다.

- 두 수의 최소공배수는 다음과 같이 구할 수 있습니다.

- 두 수를 최대공약수로 나눕니다.
- 6을 2로 나누면 3이 됩니다.
- 10을 2로 나누면 5가 됩니다.
- 최대공약수와 나눈 몫을 곱해줍니다.
- 2 × 3 = 6
- 2 × 5 = 10
- 나눈 몫을 곱한 두 수를 곱합니다.
- 6 × 10 = 60
- 최대공약수로 나눠줍니다.
- 60 / 2 = 30

따라서 6과 10의 최소공배수는 30입니다.
