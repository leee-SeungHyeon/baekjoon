```python
def solution(n, words):
    used_words = set()
    for i in range(len(words)):
        if i > 0 and (words[i-1][-1] != words[i][0] or words[i] in used_words):
            who = (i % n) + 1
            turn = (i // n) + 1
            return [who,turn]
            break
        else:
            used_words.add(words[i])
    else:
        return [0, 0]
```
