```python
n, m = map(int, input().split())

n_set = set()
m_set = set()

for i in range(n):
    n_set.add(input())
    
for i in range(m):
    m_set.add(input())

result = sorted(list(n_set & m_set))

print(len(result))
for i in result:
    print(i)
```

```python
n, m = map(int, input().split())

n_vocab = dict()
name = []
for i in range(n):
    n_vocab[input()] = 0
    
for i in range(m):
    y = input()
    if y in n_vocab:
        name.append(y)

name.sort()    
print(len(name))
for i in name:
    print(i)
```
