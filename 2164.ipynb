```python
n = int(input())

n_list = [i for i in range(n+1) if i != 0]

while len(n_list) != 1:
    
    zz = []
    del n_list[0]
    
    for i in n_list[1:]:
        zz.append(i)
    zz.append(n_list[0])
    
    n_list = zz
    
print(*n_list)
```

- 위의 코드는 시간초과가 나왔다
- 아마 리스트로 짜서 시간초과가 난것으로 생각된다

```python
from collections import deque

N = int(input())
n_list = deque([i for i in range(1, N+1)])

while(len(n_list) >1):
    n_list.popleft() 
    n_list.rotate(-1)
    
print(*n_list)
```

- deque는 시작점의 값을 넣고 빼거나, 끝 점의 값을 넣고 빼는 데 최적화된 연산 속도를 제공
- popleft는 가장 왼쪽에 있는 요소(첫번째)를 제거
- rotate(-1)는 요소들이 왼쪽으로 회전
    - 즉 a = [1,2,3]이면 a.ratate(-1)을 하면 [2,3,1]이 됨
