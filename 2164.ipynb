```python
n = int(input())

n_list = [i for i in range(n+1) if i != 0]

while len(n_list) != 1:
    
    zz = []
    del n_list[0]
    
    for i in n_list[1:]:
        zz.append(i)
    zz.append(n_list[0])
    
    n_list = zz
    
print(*n_list)
```

```python
from collections import deque

N = int(input())
n_list = deque([i for i in range(1, N+1)])

while(len(n_list) >1):
    n_list.popleft() 
    n_list.rotate(-1)
    
print(*n_list)
```
