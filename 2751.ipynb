```python
m = int(input())

num = []

for _ in range(m):
    num.append(int(input()))
    
for i in sorted(num):
    print(i)
```

- 위 코드는 시간초과가 되고, 아래 코드는 정답으로 인정해줌
- 아마 1,000,000까지 주어졌을때의 시간때문에 그런것으로 생각됨

```python
import sys
n = int(input())
l = []
for i in range(n):
    l.append(int(sys.stdin.readline()))
for i in sorted(l):
    sys.stdout.write(str(i)+'\n')
```
