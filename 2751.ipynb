```python
m = int(input())

num = []

for _ in range(m):
    num.append(int(input()))
    
for i in sorted(num):
    print(i)
```

```python
import sys
n = int(input())
l = []
for i in range(n):
    l.append(int(sys.stdin.readline()))
for i in sorted(l):
    sys.stdout.write(str(i)+'\n')
```
