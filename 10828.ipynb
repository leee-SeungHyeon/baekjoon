```python
n = int(input())

st1 = []

for _ in range(n):
    
    st = input().split()
    
    if st[0]('push'):
        st1.append(int(st[1]))
        
    elif st[0] == 'pop':
        if len(st1) == 0:
            print(-1)
        else:
            print(st1.pop())
    
    elif st[0] == 'size':
        print(len(st1))
        
    elif st[0] == 'empty':
        if len(st1) == 0:
            print(1)
        else:
            print(0)
            
    elif st[0] == 'top':
        if len(st1) == 0:
            print(-1)
        else:
            print(st1[0])
    
```
