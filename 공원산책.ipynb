```python
park = ["SOO","OOO","OOO"]
routes = 	["E 2","S 2","W 1"]	

row = len(park)
col = len(park[0])

park_map = [list(row) for row in park]

S_point = [(0, i) for i in range(len(park_map[0])) if park_map[0][i] =='S']

x = S_point[0][0]
y = S_point[0][1]


for route in routes:
    x_flag, y_flag = False, False
    
    direction, step = route.split(' ')
    step = int(step)
    if direction == 'E':
        if  y + step <= col-1:
            for i in range(y, y+step+1):
                if park_map[x][i] == 'X':
                    x_flag = True
                    break
            if not x_flag:
                y += step
                    
    elif direction == 'W':
        if  y - step >= 0:
            for i in range(y, y-step-1, -1):
                if park_map[x][i] == 'X':
                    x_flag = True
                    break
            if not x_flag:
                y -= step        
                
    elif direction == 'S':
        if  x + step <= row-1:
            for i in range(x, x+step+1):
                if park_map[i][y] == 'X':
                    y_flag = True
                    break
            if not y_flag:
                x += step    

    elif direction == 'N':
        if  x - step >= 0:
            for i in range(x, x-step-1, -1):
                if park_map[i][y] == 'X':
                    y_flag = True
                    break
            if not y_flag:
                x -= step  
    
                
print([x,y])
```

```python
park = ["SOO","OXX","OOO"]
routes = ["E 2","S 2","W 1"]

dx = {'N': -1, 'S': 1, 'E': 0, 'W':0}
dy = {'N': 0, 'S': 0, 'E': 1, 'W':-1}

x = -1

for p in park:
    x += 1
    if 'S' in p:
        y = p.index('S')
        break
    
for route in routes:
    direction , step = route.split(' ')
    for i in range(1, int(step)+1):
        nx = x + i*dx[direction]
        ny = y + i*dy[direction]
        if 0 <= nx < len(park) and 0 <= ny < len(park[0]) and park[nx][ny] != 'X':
            continue
        else:
            break
    else:
        x, y = nx, ny
    
    print(x,y)
```
