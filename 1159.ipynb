```python
n = int(input())
name = []
result = []
for _ in range(n):
    player = input()
    name.append(player[0])

first_name = set(name)   
name.sort()

for i in first_name:
    if name.count(i) > 5:
        result.append(i)
        
if len(result) > 0:
    print(''.join(sorted(result)))

else:
    print("PREDAJA")
```
