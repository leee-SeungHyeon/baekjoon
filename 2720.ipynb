```python
N = int(input())

coin_list = [25,10,5,1]
answer = []

for _ in range(N):
    
    coin = int(input())
    coin_box = [0 for _ in range(4)]
    for i in range(len(coin_list)):
        
        n = coin//coin_list[i]
        coin -= coin_list[i]* n
        coin_box[i] += n
        
    answer.append(coin_box)
    
for i in answer:
    for j in i:
        print(j, end=' ')
    print()
```
