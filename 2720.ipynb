```python
N = int(input())

coin_list = [25,10,5,1]
answer = []

for _ in range(N):
    
    coin = int(input())
    coin_box = [0 for _ in range(4)]
    for i in range(len(coin_list)):
        
        n = coin//coin_list[i]
        coin -= coin_list[i]* n
        coin_box[i] += n
        
    answer.append(coin_box)
    
for i in answer:
    for j in i:
        print(j, end=' ')
    print()
```

- 더 쉬웠던 방법이 있지만, 나는 너무 어렵게 푼 느낌이다.. 브3 문제인데.. 참 많이 부족하다..

```python
changes = [25, 10, 5, 1]
T = int(input())

for _ in range(T) :
    C = int(input())
    res = []

    for i in changes :
        res.append(C // i)	# 몫이 개수
        C = C % i	# 나머지는 다시 C에 저장
        
    print(*res)
```
