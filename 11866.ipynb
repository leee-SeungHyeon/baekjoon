```python
from collections import deque

n , k = map(int, input().split())

n_list = deque([i for i in range(1,n+1)])

result = []

while len(n_list) != 0:
    for _ in range(k-1):
        n_list.append(n_list.popleft())
        
    result.append(str(n_list.popleft()))
    
print('<'+', '.join(result)+'>')
```
