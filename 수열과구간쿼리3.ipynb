```python
def solution(arr, queries):
    
    for i in range(len(queries)):
        inx_a = queries[i][0]
        inx_b = queries[i][1]

        num_a = arr[inx_a]
        num_b = arr[inx_b]

        arr[inx_b] = num_a
        arr[inx_a] = num_b
    
    return arr
```
