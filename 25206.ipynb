```python
score = {'A+':4.5, 'A0':4.0, 'B+':3.5, 'B0':3.0, 'C+':2.5, 'C0':2.0,
         'D+':1.5, 'D0':1.0, 'F':0}

total_p = 0
pxs = 0

for i in range(20):
    _, p, s = input().split()
    p = float(p)
    
    if s != 'P':
        total_p += p
        pxs += p * score[s]
        
print('%.6f'%(pxs/total_p))
        
```
