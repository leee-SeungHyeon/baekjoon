```python
def solution(clothes):
    clothes_book = {}

    for name, type  in clothes:
        if type in clothes_book.keys():
            clothes_book[type] += [name]
        else:
            clothes_book[type] = [name]

    count = 1
    for value in clothes_book.values():
        count *= (len(value)+1)

    return count -1
```
