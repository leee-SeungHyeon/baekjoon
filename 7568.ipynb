```python
n = int(input())

result = []
humans = []

for _ in range(n):
    k, h = map(int, input().split())
    humans.append((k,h))
    
for i in range(n):
    count = 0
    for j in range(n):
        if humans[i][0] < humans[j][0] and humans[i][1] < humans[j][1]:
            count += 1
    result.append(count+1)
    
print(*result, end=' ')
```
