```python
def solution(a, b, n):
    '''
    a =        
    b =   a       
    n =      
    '''
    result = 0
    
    while n>=a:
        
        #    
        new_cola = n // a
        
        #
        result += new_cola * b
        
        #     +  
        n = n%a + new_cola * b
        
    return result
```
