```python
n,m = map(int, input().split())

monster = {}
monster_ = {}

for i in range(n):
    name = input()
    monster[i+1] = name
    monster_[name] = i+1
        
for i in range(m):
    ques = input()
    
    if ques.isnumeric():
        print(monster[int(ques)])
        
    else:
        print(monster_[ques])
```
